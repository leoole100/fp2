```python
img = rand(10, 50)*.5

x, y = np.meshgrid(*(range(s) for s in img.shape))
simga = (1, 7)
gaussian = lambda x, y: np.exp(-((x-5)**2/(2*simga[0]**2) + (y-25)**2/(2*simga[1]**2)))
img += gaussian(x, y).T
img *= 1e6

plt.imshow(img, aspect="auto", extent=(100, 500, -.5, img.shape[0]-.5))
plt.colorbar(label="counts")
plt.savefig('fig/2024-10-01 img plot.pdf')
plt.show()
```

```python
# make a 3d plot of the img
fig = plt.figure(figsize=figsize_small)
ax = fig.add_subplot(111, projection='3d')
m = ax.plot_surface(*np.meshgrid(range(img.shape[1]), range(img.shape[0])), img, cmap=cmap)
plt.show()
```
