```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def potf(x):
    return 0*x

def potpot(x):
    return 0*x

def potq(x):
    return 10*x**2/2

def step(x,pot,dx=0.01):
    dx = np.random.choice([-1,1],len(x))*dx
    return x + (np.random.random(len(x))<np.exp(-pot(x+dx)+pot(x)))*dx
```

```python
def do_sim(num_steps,num_traj,pot,xinit=0):
    x = [np.ones(num_traj)*xinit]
    for i in range(1,num_steps):
        x.append(step(x[i-1],pot))
    x = np.swapaxes(np.array(x),0,1)
    msd = np.average(x**2,axis=0)
    fig,ax = plt.subplots(2,1,figsize=(20,10))
    for j in range(0,num_traj):
        ax[0].plot(x[j])
    ax[1].plot(msd)
```

```python
do_sim(10000,50,potf)
```
